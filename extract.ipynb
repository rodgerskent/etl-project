```python
# This notebook converts the main license file from the State of Colorado into two license files
# One specific to anesthesia license information
# A second with all registered nurses for license reconcilation management
```

```python
# Dependencies
import pandas as pd
```

```python
df=pd.read_csv('license_central.csv')
df.head()
```

```python
# Consolidated list of anesthesiologist Assistants, Colorado CRNA's and Compact CRNA's
anesthesia = df[(df["subCategory"] == "CRNA") + (df["subCategory"] == "C-CRNA") + (df["licensePrefix"] == "ANT")][(df["licensePrefix"] != "MSAPN")][(df["licensePrefix"] != "MSRXN")]
anesthesia.to_csv("license_anesthesia.csv", index=False, header=True)
anesthesia.head()
```

```python
anesthesia.info()
```

```python
# Complete listing of licenses RN's (registered nurses) in State of Colorado. 
# The CRNA's are duplicative on this list and the anesthesia list because they must maintain two licenses
regnurse = df[(df["licensePrefix"] == "RN")]
regnurse.to_csv("license_regnurse.csv", index=False, header=True)
regnurse.head()
```
